```python
import csv
import pandas as pd
import os
```

```python
def get_image_set():
    image_list =os.listdir('./scraper/images')
    image_list.sort()
    image_dict = {int(img.split('-')[0]):img for img in image_list}
    return[ item[1] for item in sorted(image_dict.items())]
```

```python
get_image_set()
```

```python
dataset = pd.read_csv('./scraper/watches.csv',names=['Name','Brand','img','price'])
dataset.head()
    
```

```python
dataset['img'] = get_image_set()
```

```python
dataset.head()
```

```python
classifier_dataset = dataset[['img','Brand']]
classifier_dataset.head()
```

```python
regressor_dataset = dataset[['img', 'Brand', 'price']]
regressor_dataset.head()
```

```python
classifier_dataset.to_csv('./clf_data.csv')
```

```python
regressor_dataset.to_csv('./reg_data.csv')
```
